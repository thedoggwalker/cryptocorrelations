```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

ticker = ["BTC-USD", "ETH-USD", "ADA-USD", "XRP-USD", "SOL-USD", "DOGE-USD","TRX-USD","AVAX-USD","MATIC-USD","LTC-USD","UNI-USD","NEAR-USD","LINK-USD"]

crypto = yf.download(ticker, start = "2021-06-06", end = "2022-06-06")

crypto.to_csv("crypto.csv")

crypto = pd.read_csv("crypto.csv", header = [0,1], index_col = [0], parse_dates = [0])

close = crypto.loc[:,"Close"].copy()

norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
%matplotlib inline
plt.style.use("seaborn")
```

```python
norm.plot(figsize = (15, 8), fontsize = 13)
plt.legend(fontsize= 13)
plt.show()
```

```python
ret = close.pct_change().dropna()
ret
```

```python
ret.describe()
```

```python
summary = ret.describe().T.loc[:, ["mean","std"]] # 'T' allows us to transpose and 'loc' allows us to pick columns
summary

summary ["mean"] = summary["mean"]*252
summary ["std"] = summary["std"]*np.sqrt(252)
summary
```
